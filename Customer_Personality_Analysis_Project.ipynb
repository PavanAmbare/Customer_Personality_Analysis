```python
# lets remove ID, Year_Birth, Z_CostContact, Z_Revenue
df_reduced = df.drop(['ID', 'Year_Birth', 'Z_CostContact', 'Z_Revenue', 'Dt_Customer'], axis=1)

# rearrange the data columns
df_reduced = df_reduced[['Education', 'Marital_Status', 'Income', 'Kidhome', 'Teenhome', 'Recency', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases', 'NumWebVisitsMonth', 'age_at_enroll', 'num_children', 'num_total_purchases','total_Mnt', 'has_children', 'days_customer_for', 'accepted_any_cmp', 'AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response', 'Complain',]]
# look at one data to see how the values look like
df_reduced.iloc[1,:]
```

```python
df_reduced.info()
```

```python
from sklearn.preprocessing import OneHotEncoder

# Select the categorical columns
categorical_columns = ['Education', 'Marital_Status']

# Initialize the OneHotEncoder
encoder = OneHotEncoder(sparse=False, drop='first')

# Encode the categorical columns
encoded_cols = encoder.fit_transform(df_reduced[categorical_columns])

# Get the unique values for each categorical column
categories = encoder.categories_

# Create new column names for the encoded columns
new_columns = []
for i, category in enumerate(categories):
    for value in category[1:]:  # Exclude the first value to avoid multicollinearity
        new_columns.append(f'{categorical_columns[i]}_{value}')

# Create a DataFrame from the encoded columns
encoded_df = pd.DataFrame(encoded_cols, columns=new_columns)

# Drop the original categorical columns from the DataFrame
df_encoded = df_reduced.drop(columns=categorical_columns)

# Concatenate the encoded columns with the original DataFrame
df_encoded = pd.concat([df_encoded, encoded_df], axis=1)

# Display the first few rows of the encoded DataFrame
print(df_encoded.head())


```

```python
from sklearn.preprocessing import StandardScaler

# Selecting only the numerical columns for scaling
numerical_columns = df_encoded.select_dtypes(include=['float64', 'int64']).columns

# Initialize the StandardScaler
scaler = StandardScaler()

# Scale the numerical columns
df_scaled = scaler.fit_transform(df_encoded[numerical_columns])

# Convert the scaled data back to a DataFrame
df_scaled = pd.DataFrame(df_scaled, columns=numerical_columns)

# Concatenate the scaled numerical columns with the categorical columns
df_scaled = pd.concat([df_scaled, df_encoded.drop(columns=numerical_columns)], axis=1)

# Display the first few rows of the scaled DataFrame
print(df_scaled.head())
```
